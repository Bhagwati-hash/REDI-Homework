```python
df.Generosity.plot(kind = "hist", bins = 50, figsize =(14,14))
plt.xlabel("Generosity")
plt.ylabel("Histogram")
plt.show()
```

```python
df.plot(kind = "scatter", x= "Happiness Score",
y ="Economy (GDP per Capita)", alpha= 0.5, color = "blue")
plt.xlabel("Happiness Score")
plt.ylabel("Economy (GDP per Capita)")
plt.title("Scatter Plot of Happinesss Score and Economy")
plt.show()
```

```python
data = df[0:10]
sns.barplot(data = data, y = 'Country or region', x = 'Happiness Score')
#plot between country and happiness score
```
